```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Create a Poisson discrete random variable (rv).
# See also https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html
rv = poisson(25)

# Compute values given by Probability mass function (pmf).
pm_values = []
for num in range(0, 40):
    pm_values.append(rv.pmf(num))

plt.title('Poisson distribution example')
plt.grid(True)
plt.plot(pm_values, linewidth=2.0)
# Highlight value for 28.
plt.plot([28], [rv.pmf(28)], marker='o', markersize=6, color='red')
plt.show()
```
